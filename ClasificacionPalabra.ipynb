# Prueba de clasificación palabras moduladas

```python
import pandas as pd
import numpy as np
```

## Embbeddigns modulados

```python
#Carga de ambos datasets
df_w2v = pd.read_csv("mod_embeddings_w2v.csv")
df_fst = pd.read_csv("mod_embeddings_fst.csv")

display(df_w2v)
display(df_fst)
```

```python
#Separacion del dataset
from sklearn.model_selection import train_test_split

X_w2v = df_w2v.drop(columns=["emotion"],inplace=False)
y_w2v = df_w2v["emotion"]

X_fst = df_fst.drop(columns=["emotion"],inplace=False)
y_fst = df_fst["emotion"]

X_train_w2v, X_test_w2v, y_train_w2v, y_test_w2v = train_test_split(X_w2v,y_w2v,test_size=0.3,random_state=0)
X_train_fst, X_test_fst, y_train_fst, y_test_fst = train_test_split(X_fst,y_fst,test_size=0.3,random_state=0)
```

```python
#Importamos escaladores y hacemos fit de los escaladores
from sklearn.preprocessing import MinMaxScaler

scaler_w2v = MinMaxScaler()
scaler_fst = MinMaxScaler()

X_train_w2v = scaler_w2v.fit_transform(X_train_w2v)
X_train_fst = scaler_fst.fit_transform(X_train_fst)
```

```python
#Importamos y entrenamos los modelo de regresión logistica
from sklearn.linear_model import LogisticRegression

lr_w2v = LogisticRegression(solver = 'lbfgs',penalty='l2', C=0.1,max_iter=200)
lr_fst = LogisticRegression(solver = 'lbfgs',penalty='l2', C=0.1)

lr_w2v.fit(X_train_w2v,y_train_w2v)
lr_fst.fit(X_train_fst,y_train_fst)
```

```python
#Revisamos metricas en training
from sklearn.metrics import classification_report

y_pred_w2v = lr_w2v.predict(X_train_w2v)
y_pred_fst = lr_fst.predict(X_train_fst)

print("W2V",classification_report(y_train_w2v,y_pred_w2v))
print("FST",classification_report(y_train_fst,y_pred_fst))
```

```python
#Probamos en el testing
X_test_w2v = scaler_w2v.transform(X_test_w2v)
X_test_fst = scaler_fst.transform(X_test_fst)

y_pred_w2v = lr_w2v.predict(X_test_w2v)
y_pred_fst = lr_fst.predict(X_test_fst)

print("W2V",classification_report(y_test_w2v,y_pred_w2v))
print("FST",classification_report(y_test_fst,y_pred_fst))
```

```python
#Prueba con SVM
from sklearn.svm import SVC
svm_w2v = SVC(kernel="rbf",degree=3,C=1)
svm_fst = SVC(kernel="rbf",degree=3,C=1)

svm_w2v.fit(X_train_w2v,y_train_w2v)
svm_fst.fit(X_train_fst,y_train_fst)
```

```python
y_pred_w2v = svm_w2v.predict(X_train_w2v)
y_pred_fst = svm_fst.predict(X_train_fst)

print("W2V",classification_report(y_train_w2v,y_pred_w2v))
print("FST",classification_report(y_train_fst,y_pred_fst))
```

```python
#Probamos en el testing
y_pred_w2v = svm_w2v.predict(X_test_w2v)
y_pred_fst = svm_fst.predict(X_test_fst)

print("W2V",classification_report(y_test_w2v,y_pred_w2v))
print("FST",classification_report(y_test_fst,y_pred_fst))
```

## Embbedings no modulados

```python
#Carga de los datasets
df_w2v = pd.read_csv("embeddings_w2v.csv")
df_fst = pd.read_csv("embeddings_fst.csv")

columns = ["N"+str(i) for i in range(100)]
columns.append("emotion")

df_w2v.columns = columns
df_fst.columns = columns

display(df_w2v)
display(df_fst)
```

```python
#Separacion del dataset
from sklearn.model_selection import train_test_split

X_w2v = df_w2v.drop(columns=["emotion"],inplace=False)
y_w2v = df_w2v["emotion"]

X_fst = df_fst.drop(columns=["emotion"],inplace=False)
y_fst = df_fst["emotion"]

X_train_w2v, X_test_w2v, y_train_w2v, y_test_w2v = train_test_split(X_w2v,y_w2v,test_size=0.3,random_state=0)
X_train_fst, X_test_fst, y_train_fst, y_test_fst = train_test_split(X_fst,y_fst,test_size=0.3,random_state=0)
```

```python
w2v_spec = X_train_w2v.copy()
w2v_spec["emotion"] = y_train_w2v.values

fst_spec = X_train_fst.copy()
fst_spec["emotion"] = y_train_fst.values


display(w2v_spec)
display(fst_spec)
```

```python
from memspectrum import MESA
def max_freq_spectrum(array):
    M = MESA()
    M.solve(array)
    frecuencies,spectrum = M.spectrum(1/len(array))
    return frecuencies[np.argmax(spectrum)]

w2v_spec["max_freq"] = w2v_spec.drop(columns=["emotion"],inplace=False).apply(max_freq_spectrum,axis=1)
fst_spec["max_freq"] = fst_spec.drop(columns=["emotion"],inplace=False).apply(max_freq_spectrum,axis=1)
```

```python
display(w2v_spec)
display(fst_spec)
```

```python
emotion_list = ["anger", "anticipation", "disgust", "fear", "joy", "sadness", "surprise", "trust"]

w2v_freq = {}
for emotion in emotion_list:
    data = w2v_spec[w2v_spec["emotion"] == emotion]
    mean_freq = np.mean(data["max_freq"])
    w2v_freq[emotion] = mean_freq
w2v_freq
```

```python
fst_freq = {}
for emotion in emotion_list:
    data = fst_spec[fst_spec["emotion"] == emotion]
    mean_freq = np.mean(data["max_freq"])
    fst_freq[emotion] = mean_freq
fst_freq
```

```python
from memspectrum import MESA
def max_freq_spectrum(array):
    M = MESA()
    p,ak,ap = M.solve(array)
    return np.mean(ap)

w2v_spec["p"] = w2v_spec.drop(columns=["emotion"],inplace=False).apply(max_freq_spectrum,axis=1)
fst_spec["p"] = fst_spec.drop(columns=["emotion"],inplace=False).apply(max_freq_spectrum,axis=1)
```

```python
display(w2v_spec)
display(fst_spec)
```

```python
from matplotlib import pyplot as plt

fig,ax = plt.subplots(figsize=(10,5))

data = []
for emotion in emotion_list:
    data.append(fst_spec[fst_spec["emotion"] == emotion]["p"].values)

ax.boxplot(data)
ax.set_xticklabels(emotion_list)
```
